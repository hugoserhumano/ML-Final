```python
dots = pd.DataFrame({"zone":['AMB', 'REF', 'CON', 'NAL', 'DET', 'BAT', 'FRL', 'SAL', 'CAR', 'MAT', 'FRE', 'PEX'],
                     "x":[4, 1, 1, 3.5, 4.8, 6.2, 3, 6.3, 8.9, 8.9, 5.7, 8.9],
                     "y":[4.3, 3.75, 1.5, 3, 3, 4.3, 1, 3, 2.8, 4.3, 1, 1]})
dots
```

```python
# Create a scatter plot
fig, ax = plt.subplots(figsize=(20, 10))
for index, d in dots.iterrows():
    ax.scatter(d["x"], d["y"], label=d["zone"], s=300)
    
# Load the background image from a local file (replace 'your_image_path.jpg' with your actual file path)
image = Image.open('Untitled.jpeg')

# Display the image as the background
ax.imshow(image, extent=[0, 10, 0, 5], zorder=0)

# Set the axis labels
ax.set_xlabel('x')
ax.set_ylabel('y')

# Manually set the scale (axis limits)
ax.set_xlim(0, 10)
ax.set_ylim(0, 5)

# Add a legend
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```
